### 3.1 What is a Signal region

A SIgnal Region (SR) is a region of the phasespace where the presence of signal, as the name suggests, is *'maximal'* or at least *'optimized'*

What does this mean?

Given a certain signal process, the distibutions of the kinematic variables of such process will be different from the ones coming from bkg process. 

Take for example the following example cases:

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

x1 = np.random.normal(5, 0.5, 100)
y1 = np.random.normal(-3, 0.7, 100)

x2 = np.random.normal(1.5, 1, 10000)
y2 = np.random.normal(1.3, 2, 10000)

x = np.arange(2, 6, 0.0025)
y = 2*x - 10
ax.scatter(x2,y2,alpha=0.3,marker='.')
ax.scatter(x1,y1,alpha=0.3,marker='.')
plt.plot(x,y,c='red')
plt.show()
```

This is the ideal case we would like to be in: signal perfectly well separated from the bkground, in this case we would define a separation which would look something like the red line and would give 100% separation of bkg ans sgn.

But now take this more realistic case:

```python
fig, ax = plt.subplots()

x1 = np.random.normal(5, 1, 100)
y1 = np.random.normal(-3, 1, 100)

x2 = np.random.normal(1.5, 5, 10000)
y2 = np.random.normal(1.3, 2, 10000)

x = np.arange(1, 9, 0.0025)
y = -0.4*(x-5)**2 - 0.8
ax.scatter(x2,y2,alpha=0.3,marker='.')
ax.scatter(x1,y1,alpha=0.3,marker='.')
plt.plot(x,y,c='red')
plt.show()
```

This is a more general case: in this case we would define a separation which would look something like the red line and would give a good separation between bkg and sgn.

### 3.2 SR definition criteria

But how do we choose the boundary of the SR?

To do so we generally use, depending on the goal, three different figures of merit:

$$ \frac{S}{B} \hspace{5mm};\hspace{5mm} \frac{S}{S + B} \hspace{5mm};\hspace{5mm} \frac{S}{\sqrt{B}}$$

The first two are almost equivalent and they become approximately the same in the limit of large bkg contribution. They are generally used when it is known from the strat that the outcome of the analysis will not end in a discovery or measuremnet but in a limit setting.

The third one is generally used when the goal is the measuremnet of a quantity (e.g. a cross section) or its discovery. This because it will be carachteristic of a region where we expect the signal to be larger or comparable to the statistical oscillations of the background.

For this reasons, while the first two are generally referred to as **'significance'**, the third one is referred to as **'discovery significance'** 

But why is there a square root at the denominator of the discovery significance? 

This is because a counting experiment follows a Poisson distribution. Therefore if I expect a number $B_{exp}$ background events, I already know that the error on this value is $\sqrt{B_{exp}}$ and the fluctuations of the background will be of the same order.
